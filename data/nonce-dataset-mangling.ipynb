```python
import pandas as pd
import numpy as np

# generate a dataset that uses the same format as the lgd_dataset.tsv from
# the generated.tab file, which is quite differently structured. 
df = pd.read_table('generated.tab')
df = df[(df['type'] == 'generated')]

maxlen =  int(len(df. index))+100
columns = ['full_sentence', 'verb_masked_sentence', 'verb_true_label', 'verb_false_label']
nonce_df = pd.DataFrame(np.full((maxlen, len(columns)), np.nan), columns = columns )
print('MAXLEN', maxlen)
for index, row in df.iterrows():
    full_sentence = row['sent']
    verb_true_label = None
    verb_false_label = None
    i = index
    # every two rows describe the same sentence
    i = (index - 1) if i % 2 == 1 else index


    if (row['class'] == 'correct'):
        verb_true_label = row['form']
        # replace the word at prefix_len with the true label since
        # the verb seems to be different than is listed
        verb_index = row['len_prefix']
        full_sentence_list = full_sentence.split()
        full_sentence_list[verb_index] = verb_true_label
        full_sentence = " ".join(full_sentence_list)

        # also generated the masked sentence at the same go
        masked_sentence_list = full_sentence_list
        masked_sentence_list[verb_index] = '[MASK]'
        verb_masked_sentence = " ".join(masked_sentence_list)

        nonce_df.loc[nonce_df.index[i], 'verb_masked_sentence'] = verb_masked_sentence
        nonce_df.loc[nonce_df.index[i], 'full_sentence'] = full_sentence
        nonce_df.loc[nonce_df.index[i], 'verb_true_label'] = verb_true_label

    elif (row['class'] == 'wrong'):
        verb_false_label = row['form']
        nonce_df.loc[nonce_df.index[i], 'verb_false_label'] = verb_false_label


nonce_df = nonce_df.dropna()
nonce_df
nonce_df.to_csv('nonce_dataset.csv')
```
